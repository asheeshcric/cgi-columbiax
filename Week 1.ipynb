# Kinematics

1. Configuration:
    - e.g: Heights of two elevators describe a configuration in space
    - e.g: Velocities of two elevators also describe a configuration in terms of velocity
    
2. Trajectory:
    - Change in configuration with respect to time
    - A curve path traced when the configuration of a system is plotted with explicit time

3. Velocity:
    - Important in describing the next state of a configuration space
    
4. State:
    - The complete configuration vector (e.g. containing config space and config velocity) that can describe the current position and the next immediate motion of a system
